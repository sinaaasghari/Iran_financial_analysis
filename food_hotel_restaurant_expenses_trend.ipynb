```python
years = [98, 99, 1400, 1401]
urban_total_value = []
rural_total_value = []

urban_data_all_years = {}
rural_data_all_years = {}
files_U = {
    98: '/content/drive/MyDrive/prj/U98.xlsx',
    99: '/content/drive/MyDrive/prj/U99.xlsx',
    1400: '/content/drive/MyDrive/prj/U1400.xlsx',
    1401: '/content/drive/MyDrive/prj/U1401.xlsx'
}

files_R = {
    98: '/content/drive/MyDrive/prj/R98.xlsx',
    99: '/content/drive/MyDrive/prj/R99.xlsx',
    1400: '/content/drive/MyDrive/prj/R1400.xlsx',
    1401: '/content/drive/MyDrive/prj/R1401.xlsx'
}
for year, path in files_U.items():
    temp_urban_data = pd.read_excel(path, sheet_name=f'U{year}P3S11') 
    urban_data_all_years[year] = temp_urban_data
for year, path in files_R.items():
    temp_rural_data = pd.read_excel(path, sheet_name=f'R{year}P3S11') 
    rural_data_all_years[year] = temp_rural_data

def convert_to_numeric(data, col):
    data[col] = pd.to_numeric(data[col], errors='coerce').fillna(0)

def aggregate_value(data):
    convert_to_numeric(data, 'value')
    total_value = data['value'].sum()
    return total_value

for year in years:

        aggregate_values = aggregate_value(urban_data_all_years[year])
        urban_total_value.append(aggregate_values)
        aggregate_values = aggregate_value(rural_data_all_years[year])
        rural_total_value.append(aggregate_values)
data = list(zip(urban_total_value, rural_total_value))
dim = len(data[0])
w = 0.6  
dimw = w / dim  

fig, ax = plt.subplots()
x = np.arange(len(data))
categories = ['Urban Expenses', 'Rural Expenses']
colors = ['blue', 'green']
for i in range(len(data[0])):
    y = [d[i] for d in data] 
    ax.bar(x + i * dimw, y, dimw, label=categories[i], color=colors[i])  # Plot each bar group

ax.set_xticks(x + dimw * (dim / 2))  
ax.set_xticklabels([str(year) for year in years])
ax.set_ylabel('Amount')
ax.set_xlabel('Year')
ax.set_title('Food, Hotels, Restaurant Expenses')
ax.legend()
plt.show()
```
