```python
all_dfs={'U98_dfs':U98_dfs,'U99_dfs':U99_dfs,'U1400_dfs':U1400_dfs,'U1401_dfs':U1401_dfs,'R98_dfs':R98_dfs,'R99_dfs':R99_dfs,'R1400_dfs':R1400_dfs,'R1401_dfs':R1401_dfs}
```

Correcting the Types

```python
for k,v in all_dfs.items():
    x = v[2]
    y = v[1]
    
    x.loc[x['income_pension'] == '\n','income_pension'] = 0
    x.loc[x['income_rent'] == '\n','income_rent'] = 0
    x.loc[x['income_interest'] == '\n','income_interest'] = 0
    x.loc[x['income_resale'] == '\n','income_resale'] = 0
    x.loc[x['income_transfer'] == '\n','income_transfer'] = 0
    x['income_pension'] = x['income_pension'].astype('float')
    x['income_rent'] =x['income_rent'].astype('float')
    x['income_interest'] =x['income_interest'].astype('float')
    x['income_resale'] =x['income_resale'].astype('float')
    x['income_transfer'] =x['income_transfer'].astype('float')

    y.loc[y['income_s_y'] == '              ','income_s_y'] = 0
    y['income_s_y'] = y['income_s_y'].astype('float')
```

Merging Each Year Data

```python
merged_dfs = {}
for k,v in all_dfs.items():
    merged_df = v[0]
    for df in v[1:]:
        merged_df = pd.merge(merged_df,df, on=['Address','member'], how='outer')
    merged_df.fillna(0,inplace=True)
    merged_dfs[k]=merged_df
```

```python
final_dfs = []
for k,v in merged_dfs.items():
    v['Total_Income']=v.iloc[:,2:].sum(axis = 1)
    final_df = v.drop(columns=['netincome_w_y', 'income_s_y','income_pension', 'income_rent', 'income_interest','income_aid', 'income_resale', 'income_transfer','subsidy'])
    final_dfs.append(final_df)
```

The Inflation Rate

```python
# Based on Data from 'https://www.cbi.ir/Inflation/Inflation_FA.aspx'
# We choose 1398 as our base year

inflation_99 =0.471
inflation_1400 =0.462
inflation_1401 =0.465
```

```python
#year 1398
final_dfs[0]['True_Income'] = final_dfs[0]['Total_Income']
final_dfs[4]['True_Income'] = final_dfs[4]['Total_Income']

#year 1399
final_dfs[1]['True_Income'] = final_dfs[1]['Total_Income'] / (1+inflation_99)
final_dfs[5]['True_Income'] = final_dfs[5]['Total_Income'] / (1+inflation_99)

#year 1400
final_dfs[2]['True_Income'] = final_dfs[2]['Total_Income'] / ((1+inflation_99)*(1+inflation_1400))
final_dfs[6]['True_Income'] = final_dfs[6]['Total_Income'] / ((1+inflation_99)*(1+inflation_1400))

#year 1401
final_dfs[3]['True_Income'] = final_dfs[3]['Total_Income'] / ((1+inflation_99)*(1+inflation_1400)*(1+inflation_1401))
final_dfs[7]['True_Income'] = final_dfs[7]['Total_Income'] / ((1+inflation_99)*(1+inflation_1400)*(1+inflation_1401))
```

The Plot based on Average of Each Year

```python
U_yearly_mean_nominal_incomes ={}
R_yearly_mean_nominal_incomes ={}
U_yearly_mean_true_incomes ={}
R_yearly_mean_true_incomes ={}

year = 1398
for df in final_dfs[0:4]:
    U_yearly_mean_nominal_incomes[year] = df['Total_Income'].mean()
    U_yearly_mean_true_incomes[year] = df['True_Income'].mean()
    year+=1

year = 1398
for df in final_dfs[4:]:
    R_yearly_mean_nominal_incomes[year] = df['Total_Income'].mean()
    R_yearly_mean_true_incomes[year] = df['True_Income'].mean()
    year+=1
```

```python
U_nominal_income_df = pd.DataFrame(list(U_yearly_mean_nominal_incomes.items()), columns=['Year', 'Total_Income'])
R_nominal_income_df = pd.DataFrame(list(R_yearly_mean_nominal_incomes.items()), columns=['Year', 'Total_Income'])
U_true_income_df = pd.DataFrame(list(U_yearly_mean_true_incomes.items()), columns=['Year', 'True_Income'])
R_true_income_df = pd.DataFrame(list(R_yearly_mean_true_incomes.items()), columns=['Year', 'True_Income'])
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 12)) 

axs[0].plot(U_nominal_income_df['Year'], U_nominal_income_df['Total_Income'], label='Total Income', marker='o')
axs[0].plot(U_true_income_df['Year'], U_true_income_df['True_Income'], label='True Income', marker='x')
axs[0].set_title('Urban Incomes')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Income')
axs[0].legend()
axs[0].grid()
axs[0].set_xticks(U_nominal_income_df['Year'])
axs[0].set_xticklabels(U_nominal_income_df['Year'].astype(int))

for i, row in U_nominal_income_df.iterrows():
    axs[0].text(row['Year'], row['Total_Income'], f"{row['Total_Income']:.0f}", fontsize=9, ha='center', va='bottom')
    
for i, row in U_true_income_df.iterrows():
    axs[0].text(row['Year'], row['True_Income'], f"{row['True_Income']:.0f}", fontsize=9, ha='center', va='bottom')

axs[1].plot(R_nominal_income_df['Year'], R_nominal_income_df['Total_Income'], label='Total Income', marker='o')
axs[1].plot(R_true_income_df['Year'], R_true_income_df['True_Income'], label='True Income', marker='x')
axs[1].set_title('Rural Incomes')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Income')
axs[1].legend()
axs[1].grid()
axs[0].set_xticks(R_nominal_income_df['Year'])
axs[0].set_xticklabels(R_nominal_income_df['Year'].astype(int))

for i, row in R_nominal_income_df.iterrows():
    axs[1].text(row['Year'], row['Total_Income'], f"{row['Total_Income']:.0f}", fontsize=9, ha='center', va='bottom')
    
for i, row in R_true_income_df.iterrows():
    axs[1].text(row['Year'], row['True_Income'], f"{row['True_Income']:.0f}", fontsize=9, ha='center', va='bottom')

plt.tight_layout() 
plt.show()
```
