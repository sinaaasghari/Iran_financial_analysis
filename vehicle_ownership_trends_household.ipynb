```python
urban_data_all_years = {}
rural_data_all_years = {}
files_U = {
    98: '/content/drive/MyDrive/prj/U98.xlsx',
    99: '/content/drive/MyDrive/prj/U99.xlsx',
    1400: '/content/drive/MyDrive/prj/U1400.xlsx',
    1401: '/content/drive/MyDrive/prj/U1401.xlsx'
}

files_R = {
    98: '/content/drive/MyDrive/prj/R98.xlsx',
    99: '/content/drive/MyDrive/prj/R99.xlsx',
    1400: '/content/drive/MyDrive/prj/R1400.xlsx',
    1401: '/content/drive/MyDrive/prj/R1401.xlsx'
}
for year, path in files_U.items():
    temp_urban_data = pd.read_excel(path, sheet_name=f'U{year}P2') 
    urban_data_all_years[year] = temp_urban_data

for year, path in files_R.items():
    temp_rural_data = pd.read_excel(path, sheet_name=f'R{year}P2') 
    rural_data_all_years[year] = temp_rural_data

vehicle_columns = ['vehicle', 'motorcycle', 'bicycle', 'car']

def count_vehicle_ownership(data, vehicle_col):
    return data[vehicle_col].sum()

def plot_vehicle_ownership(year, vehicle_counts, data_type):
    vehicle_types = list(vehicle_counts.keys())
    counts = list(vehicle_counts.values())
    
    fig, ax = plt.subplots(figsize=(8, 6))  # Larger plot size
    bars = ax.bar(vehicle_types, counts, color='skyblue' if data_type == 'Urban' else 'lightgreen', width=0.6)

    
    plt.title(f'Vehicle Ownership in {data_type} - Year {year}', fontsize=16)
    plt.xlabel('Vehicle Type', fontsize=14)
    plt.ylabel('Number of Households', fontsize=14)
    
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    
    plt.tight_layout() 
    plt.show()

for year in [98, 99, 1400, 1401]:
    urban_vehicle_counts = {}
    rural_vehicle_counts = {}

    for vehicle in vehicle_columns:
        if vehicle in urban_data_all_years[year].columns:
            count = count_vehicle_ownership(urban_data_all_years[year], vehicle)
            urban_vehicle_counts[vehicle] = count
    for vehicle in vehicle_columns:
        if vehicle in rural_data_all_years[year].columns:
            count = count_vehicle_ownership(rural_data_all_years[year], vehicle)
            rural_vehicle_counts[vehicle] = count
    plot_vehicle_ownership(year, urban_vehicle_counts, 'Urban')
    plot_vehicle_ownership(year, rural_vehicle_counts, 'Rural')
```
